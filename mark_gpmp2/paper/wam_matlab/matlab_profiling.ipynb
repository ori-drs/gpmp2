# Plots

```python
font = {'family' : 'normal',
        'weight' : 'normal',
        'size'   : 16}

matplotlib.rc('font', **font)

x = ["Publishing Trajectory", "Object Tracking",
"Re-formatting SDFs", "Updating Factors",
"Predicting Composite SDFs", "Updating Occupancy Grid",
"Receiving Joint Feedback", "Re-optimising",
"Adding Configuration Priors", "Other" , "Self-time"]

time = [27.57, 13.52, 13.46, 10.17, 8.11, 5.06, 3.76, 3.23, 2.59, 1.59, 10.93]

x_pos = [i for i, _ in enumerate(x)]
time.reverse()
x.reverse()


fig, ax = plt.subplots(figsize=(10, 6), dpi=90,)    
width = 0.75 # the width of the bars 
# ind = np.arange(len(y))  # the x locations for the groups
# ax.barh(ind, y, width, color="blue")
# ax.set_yticks(ind+width/2)
# ax.set_yticklabels(x, minor=False)

plt.barh(x_pos, time, color='tab:blue')
# plt.ylabel("Function")
plt.xlabel("Time (% of Update Loop)")
plt.yticks(x_pos, x)
plt.xlim([0,33])
# plt.title('Profiling')
for i, v in enumerate(time):
    ax.text(v+0.5 , i-0.2 , "{0:.1f}%".format(v), color='black')
    
plt.tight_layout()
plt.savefig('matlab_profiling.png', dpi=300)
plt.show()
```
